```python
m=int(input("enter marks: "))
if  m<25:
  print("grade-F")
elif 25<m<=45:
  print("grade-E")
elif 45<m<=50:
    print("grade-D")
elif 50<m<=60:
   print("grade-c")
elif 60<m<=80:
    print("grade-B")
else: 
     print("grade-A")
```

## Class Activity:

Get a minimum number and maximum number from user, and print all the odd numbers between the minimum and maximum number

`while` loop

```python
x=int(input("Enter the minimum limit: "))
y=int(input("Enter the maximum limit: "))

while(x<=y):
  if (x%2 == 0):
    print(x+1,end=" ")
    x+=2
  else :
    print(x,end=" ")
    x+=2
```
